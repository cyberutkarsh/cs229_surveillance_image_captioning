```python
# import pandas
import pandas as pd
```

```python
# read csv file in pandas dataframe
df = pd.read_csv('../experiment_2_annotations.csv')
```

```python
df.head()
```

```python
annotation_dict_list = []
```

```python
# iterate over rows in dataframe
for index, row in df.iterrows():
    filename = row['file name']
    
    # create an array of human captions that are not empty
    human_captions = [row['human caption 1'], row['human caption 2'], row['human caption 3'], row['human caption 4'], row['human caption 5']]
    human_captions_non_blank = [caption for caption in human_captions if str(caption) != '' and str(caption) != 'nan']

    # loop over human_captions_non_blank
    for caption in human_captions_non_blank:
        annotation_dict = {}
        # create a dictionary with filename and caption
        annotation_dict = {'filename': filename, 'caption': caption, 'id': index+829716, 'image_id': index+581922}
        # append the dictionary to annotation_dict_list
        annotation_dict_list.append(annotation_dict)
```

```python
annotation_dict_list
```

```python
df_annotations = pd.DataFrame(annotation_dict_list)
```

```python
df_annotations.head(10)
```

```python
#rename files in python
import os
for index, row in df_annotations.iterrows():
    filename = row['filename']
    # get file extension
    file_extension = filename.split('.')[-1]
    imageid = int(row['image_id']).__format__('012')
    new_filename = 'COCO_train2014_' + str(imageid) + '.' + file_extension
    path = '../all_images/'
    # print(filename)
    # print(new_filename)    
    # break
    try:
        os.rename(path + filename, path + new_filename)
    except:
        print('error as already renamed')
```

```python
df_annotations.to_json('train_caption.json',orient='records')
```
